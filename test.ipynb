```python
import cv2
import numpy as np
from random import *
from tkinter import filedialog

def create_image(height, width):
    img = np.zeros((height,width,3), np.uint8)
    img[:,:] = [randint(0, 256),randint(0, 256),randint(0, 256)]
    return img

def generate_rectangle(img):
    height, width = img.shape[:2]
    # generate an image and a random rectangle
    p1 = [randint(0, height), randint(0, width)]
    p2 = [randint(0, height), randint(0, width)]
    p3 = [randint(0, height), randint(0, width)]
    p4 = [randint(0, height), randint(0, width)]
    cv2.drawContours(img, [np.array([p1,p2,p3,p4])], 0, (randint(0, 256),randint(0, 256),randint(0, 256)), -1)
    return img

def generate_triangle(img):
    height, width = img.shape[:2]
    # generate an image and a random rectangle
    p1 = [randint(0, height), randint(0, width)]
    p2 = [randint(0, height), randint(0, width)]
    p3 = [randint(0, height), randint(0, width)]
    cv2.drawContours(img, [np.array([p1,p2,p3])], 0, (randint(0, 256),randint(0, 256),randint(0, 256)), -1)
    return img

def generate_ellipse(img):
    height, width = img.shape[:2]
    # generate an image and a random rectangle
    image = cv2.ellipse(img, [randint(0, height), randint(0, width)], [randint(0, height), randint(0, width)], randint(0, 90), 0, 360, randint(0, width), -1)
    return img

def generate_star(image):
    height, width = img.shape[:2]
    center = [randint(0, height), randint(0, width)]
    num_points = randint(4, 6)
    radius = randint(0, width/2)
    
    # Generate random angles for the star points
    angles = [2 * np.pi * i / num_points for i in range(num_points)]

    # Generate random lengths for the star points
    lengths = [np.random.uniform(0.5, 1.0) * radius for _ in range(num_points)]

    # Generate star points
    points = []
    for angle, length in zip(angles, lengths):
        x = int(center[0] + length * np.cos(angle))
        y = int(center[1] + length * np.sin(angle))
        points.append((x, y))
        x = int(center[0] + length/3 * np.cos(angle + np.pi / num_points))
        y = int(center[1] + length/3 * np.sin(angle + np.pi / num_points))
        points.append((x, y))

    # Convert the points to NumPy array format
    points = np.array(points, np.int32)

    # Create a blank mask with the same size as the image


    # Draw the filled star on the mask
    cv2.fillPoly(img, [points], color=[randint(0, height), randint(0, width)])

    # Bitwise AND the image and mask to fill the star on the image
    return image
    
    


img = create_image(512, 512)

generate_triangle(img)

generate_rectangle(img)

generate_ellipse(img)

generate_star(img)

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
#add additive gaussian noise to the image
def additive_noise(img):
    height, width = img.shape[:2]
    noise = np.zeros((height,width,3), np.uint8)
    mean = np.array([0, 0, 0])
    sigma = np.array([0.2,0.2,0.2])
    cv2.randn(noise, mean, sigma)
    img = cv2.add(img, noise)
    return img

def multiplicative_noise(img):
    height, width = img.shape[:2]
    noise = np.zeros((height,width,3), np.uint8)
    mean = np.array([1, 1, 1])
    sigma = np.array([0.2,0.2,0.2])
    cv2.randn(noise, mean, sigma)
    img = cv2.multiply(img, noise)
    return img

img_add = img.copy()
img_mul = img.copy()

additive_noise(img_add)
multiplicative_noise(img_mul)

cv2.imshow('additive',img_add)
cv2.imshow('multiplicative',img_mul)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
